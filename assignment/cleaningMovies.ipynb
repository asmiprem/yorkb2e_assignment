```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movies.csv")
display(df)
```

# Clean and transform the above dataset to add the following columns: Start year, end year, director, actors.

```python
df[['start_year', 'end_year']] = df['YEAR'].str.strip('()').str.split('–', expand=True)
df.head()
```

# Clean and transform the above dataset to add the following columns:director, actors.

```python
df[['Director', 'Actors']] = df['STARS'].str.split('\n\|\n|Stars:|Star:', expand=True)
df['Director'] = df['Director'].str.replace('Director:', '').str.strip()
df['Director'] = df['Director'].str.replace('Directors:', '').str.strip()
df['Actors'] = df['Actors'].str.replace('Stars:', '').str.strip()
df['Director'] = df['Director'].str.replace('[\n,\|]', '',regex=True).str.strip()
df['Actors'] = df['Actors'].str.replace('\n', '')
df[['Director', 'Actors']]






df.head() 

```

# converting Gross and Votes to numeric

```python
df['Gross'].dtype
df['VOTES'].dtype
df['VOTES'] = df['VOTES'].str.replace(',', '')
df['Votes_Numeric']=pd.to_numeric(df['VOTES'])
df['Gross'] = pd.to_numeric(df['Gross'].str.replace('[\$,M]', '', regex=True), errors='coerce')

```

# create a new csv that contains a subset of the original data where "Gross" is a known value, and all other numerical statistics are present


```python
selected_df = df.dropna(subset=['Gross']).select_dtypes(include='number')


selected_df.to_csv('SubsetMoviesData.csv', index=False)
```

```python
description_df=df.dropna(subset=['RATING', 'ONE-LINE','Gross'])


description_df['Description Length'] =description_df['ONE-LINE'].apply(len)


plt.figure(figsize=(10, 6))
bar_colors = 'orange'
plt.bar(description_df['RATING'],description_df['Description Length'],color=bar_colors)


plt.xlabel('RATING')
plt.ylabel('Description Length')
plt.title('Correlation between Rating and Description Length')


plt.show()
```

```python
selected_df = df.dropna(subset=['RATING','RunTime','Gross']).select_dtypes(include='number')

plt.figure(figsize=(10, 6))
bar_colors = 'pink'
plt.bar(selected_df['RATING'], selected_df['RunTime'],color=bar_colors)


plt.xlabel('RATING')
plt.ylabel('RunTime')
plt.title('Bar Plot of RATING vs RunTime')

```

```python



counts,bins,patches=plt.hist(selected_df['RunTime'], color='salmon', edgecolor='black')
plt.xlabel('RunTime')
plt.ylabel('Frequency')
plt.title('Histogram of RunTime')
plt.xticks(bins)  


plt.show()
```
