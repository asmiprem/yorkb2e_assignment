# Data Analysis Report
 
## Introduction

The dataset under consideration contains information about students. It includes various attributes related to the students' personal, family, and academic background. The goal is to perform data analysis and build predictive models for the final grade (G3) based on the available features.

## Exploratory Data Analysis (EDA)

- Descriptive Statistics:
The dataset consists of both numerical and categorical features. Descriptive statistics were computed to understand the distribution and characteristics of the numerical variables. Additionally, histograms were created to gain insight into various numerical variables, illustrating the frequency distribution of the data within specified intervals. This visual representation shows the shape and spread of the data.

The distribustion of G3 shows some values in the range 1-4 are missing.
- Correlation Analysis:
A correlation matrix was generated to identify relationships between numerical features. Heatmaps were used for visual representation, providing insights into potential correlations.

- Categorical Feature Visualization:
Categorical features were visualized using count plots and box plots to observe their distribution and potential impact on the target variable (G3).

## Feature Engineering and Feature Selection:

Several feature engineering techniques were applied to create new variables and subsets of the data. This included the addition of alcohol-related features and the combination of parental education levels, as well as the combination of health and age. The features selected for model evaluation were based on Mutual Information scores, the correlation matrix, and visualizations of categorical features. Notably, features such as 'G1,' 'G2,' 'absences,' and 'Failures' demonstrated high Mutual Information scores, indicating their importance in the modeling process.

   

## Model Building:

- Data Preprocessing
The dataset was divided into training and testing sets, and I opted for one-hot encoding for categorical variables during preprocessing. This method was chosen to ensure proper representation of non-ordinal categories, preventing misinterpretation and bias in model training, ultimately enhancing the model's ability to handle categorical data effectively.

- Model Selection
Four regression models were chosen for evaluation because they perform well with small datasets and effectively address overfitting.

    - Random Forest Regressor
    - Decision Tree Regressor
    - Ridge Regressor
    - Linear Regression

- Model Evaluation
Models were evaluated for different feature set using cross-validation mean squared error (CV MSE), mean squared error (MSE), R2 score, and mean absolute error (MAE) on both the training and testing sets.

## Model Comparison and Selection

The performance of each model was compared based on CV MSE, MSE, R2 score, and MAE. The results were visualized through bar plots to provide a clear understanding of each model's strengths and weaknesses.


After evaluating eight different feature sets, the CV MSE, MSE, R2 score, and MAE revealed that the Random Forest model consistently outperformed other models across the varied feature sets.

## Hyperparameter Tuning
A grid search was conducted to determine optimal hyperparameters for the Random Forest Regressor. The best model was selected, and its performance metrics on the test set were analyzed.

## Findings and Analysis:
The analysis unveils insights into the factors influencing students' final grades. The Random Forest Regressor, with tuned hyperparameters, demonstrated the best overall performance. Previous grades, absences, students' intention to pursue higher studies,failures, study time, and features such as alcohol, along with social aspects like parents' education parents job and family relationships, including going out, were identified as crucial elements contributing to the prediction of final grades.
